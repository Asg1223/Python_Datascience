```python
import numpy as np
from scipy import stats
# ä¸ãˆã‚‰ã‚ŒãŸãƒ‡ãƒ¼ã‚¿ã€€(é©å®œå¤‰æ›´)
data = np.array([12.1,11.5,10.9,12.3,11.8,12.0,11.6,12.4,11.2,11.9])

# æ¨™æœ¬ã‚µã‚¤ã‚º
n = len(data)
xbar = np.mean(data)
s = np.std(data, ddof=1)
df = n - 1
t_crit = stats.t.ppf(1 - 0.05, df)  # 90%ä¿¡é ¼åŒºé–“ â†’ ä¸¡å´5% = 0.05
margin = t_crit * s / math.sqrt(n)
ci_mean = (xbar - margin, xbar + margin)
print(f"æ¨™æœ¬å¹³å‡ = {xbar:.3f}, æ¨™æœ¬æ¨™æº–åå·® = {s:.3f}")
print(f"tå€¤ = {t_crit:.3f}")
print(f"90%ä¿¡é ¼åŒºé–“: {ci_mean[0]:.3f} ï½ {ci_mean[1]:.3f}\n")
```

```python
import numpy as np
from scipy.stats import chi2
# ä¸ãˆã‚‰ã‚ŒãŸãƒ‡ãƒ¼ã‚¿
ph_values = np.array([12.1,11.5,10.9,12.3,11.8,12.0,11.6,12.4,11.2,11.9])

# æ¨™æœ¬ã‚µã‚¤ã‚º
n = len(ph_values)

# æ¨™æœ¬å¹³å‡ã‚’è¨ˆç®—
sample_mean_ph = np.mean(ph_values)

# æ¨™æœ¬åˆ†æ•£ã‚’è¨ˆç®— (ä¸ååˆ†æ•£ã‚’ä½¿ç”¨ã™ã‚‹ãŸã‚ ddof=1 ã‚’æŒ‡å®š)
sample_variance = np.var(ph_values, ddof=1)
# è‡ªç”±åº¦
degrees_of_freedom_ph = n - 1

# ä¿¡é ¼åº¦95%ã®å ´åˆã€ä¸¡å´ç¢ºç‡ãŒ5%ãªã®ã§ã€ç‰‡å´ç¢ºç‡ã¯2.5% = 0.025
# ä¸‹å´ã‚«ã‚¤äºŒä¹—å€¤ (ç´¯ç©ç¢ºç‡ 0.025) ã¨ ä¸Šå´ã‚«ã‚¤äºŒä¹—å€¤ (ç´¯ç©ç¢ºç‡ 0.975) ã‚’ chi2.ppf() ã§æ±‚ã‚ã‚‹
chi2_lower = chi2.ppf((1 - 0.95) / 2, degrees_of_freedom_ph)
chi2_upper = chi2.ppf(1 - (1 - 0.95) / 2, degrees_of_freedom_ph)

# æ¯åˆ†æ•£ã®ä¿¡é ¼åŒºé–“ã‚’è¨ˆç®—
lower_bound_var = (degrees_of_freedom_ph * sample_variance) / chi2_upper
upper_bound_var = (degrees_of_freedom_ph * sample_variance) / chi2_lower

print(f"æ¯åˆ†æ•£ã®95%ä¿¡é ¼åŒºé–“ã¯ [{lower_bound_var:.3f}, {upper_bound_var:.4f}] ã§ã™ã€‚")
```

```python
import math
from scipy.stats import norm

n = 200       # æ¨™æœ¬ã‚µã‚¤ã‚º
x = 118       # æˆåŠŸ
p0 = 0.5      # ä»®å®šã™ã‚‹æ¯æ¯”ç‡ï¼ˆå¸°ç„¡ä»®èª¬ï¼‰
alpha = 0.05  # æœ‰æ„æ°´æº–5%

phat = x / n
z = (phat - p0) / math.sqrt(p0 * (1 - p0) / n)

p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"æ¨™æœ¬æ¯”ç‡ phat = {phat:.3f}")
print(f"æ¤œå®šçµ±è¨ˆé‡ z = {z:.3f}")
print(f"på€¤ = {p_value:.4f}")

if p_value < alpha:
    print("â‡’ å¸°ç„¡ä»®èª¬ã‚’æ£„å´ï¼ˆæœ‰æ„å·®ã‚ã‚Šï¼‰")
else:
    print("â‡’ å¸°ç„¡ä»®èª¬ã‚’æ£„å´ã§ããªã„ï¼ˆæœ‰æ„å·®ãªã—ï¼‰")
```

```python
import math

mu_0 = 100       # å¸°ç„¡ä»®èª¬ã§ã®æ¯å¹³å‡
sigma = 3.6     # æ¯æ¨™æº–åå·® (æ—¢çŸ¥)
n = 12          # æ¨™æœ¬ã‚µã‚¤ã‚º
x_bar = 102.3     # æ¨™æœ¬å¹³å‡
alpha = 0.05     # æœ‰æ„æ°´æº–

z_stat = (x_bar - mu_0) / (sigma / math.sqrt(n))

p_value = 2 * (1 - 0.5 * (1 + math.erf(abs(z_stat) / math.sqrt(2))))


# æ¤œå®šçµæœ
if p_value < alpha:
    print(f"æœ‰æ„æ°´æº– {alpha} ã§å¸°ç„¡ä»®èª¬ã‚’æ£„å´ã™ã‚‹ã€‚")
else:
    print(f"æœ‰æ„æ°´æº– {alpha} ã§å¸°ç„¡ä»®èª¬ã‚’æ£„å´ã—ãªã„ã€‚")
```
