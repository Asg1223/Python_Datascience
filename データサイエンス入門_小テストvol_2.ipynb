問１

```python
import numpy as np
from scipy import stats
# 与えられたデータ　(適宜変更)
data = np.array([12.1,11.5,10.9,12.3,11.8,12.0,11.6,12.4,11.2,11.9])

# 標本サイズ
n = len(data)
xbar = np.mean(data)
s = np.std(data, ddof=1)
df = n - 1
t_crit = stats.t.ppf(1 - 0.05, df)  # 90%信頼区間 → 両側5% = 0.05
margin = t_crit * s / math.sqrt(n)
ci_mean = (xbar - margin, xbar + margin)
print(f"標本平均 = {xbar:.3f}, 標本標準偏差 = {s:.3f}")
print(f"t値 = {t_crit:.3f}")
print(f"90%信頼区間: {ci_mean[0]:.3f} ～ {ci_mean[1]:.3f}\n")
```

```python
import numpy as np
from scipy.stats import chi2
# 与えられたデータ
ph_values = np.array([12.1,11.5,10.9,12.3,11.8,12.0,11.6,12.4,11.2,11.9])

# 標本サイズ
n = len(ph_values)

# 標本平均を計算
sample_mean_ph = np.mean(ph_values)

# 標本分散を計算 (不偏分散を使用するため ddof=1 を指定)
sample_variance = np.var(ph_values, ddof=1)
# 自由度
degrees_of_freedom_ph = n - 1

# 信頼度95%の場合、両側確率が5%なので、片側確率は2.5% = 0.025
# 下側カイ二乗値 (累積確率 0.025) と 上側カイ二乗値 (累積確率 0.975) を chi2.ppf() で求める
chi2_lower = chi2.ppf((1 - 0.95) / 2, degrees_of_freedom_ph)
chi2_upper = chi2.ppf(1 - (1 - 0.95) / 2, degrees_of_freedom_ph)

# 母分散の信頼区間を計算
lower_bound_var = (degrees_of_freedom_ph * sample_variance) / chi2_upper
upper_bound_var = (degrees_of_freedom_ph * sample_variance) / chi2_lower

print(f"母分散の95%信頼区間は [{lower_bound_var:.3f}, {upper_bound_var:.4f}] です。")
```

問3

```python
import math
from scipy.stats import norm

n = 200       # 標本サイズ
x = 118       # 成功
p0 = 0.5      # 仮定する母比率（帰無仮説）
alpha = 0.05  # 有意水準5%

phat = x / n
z = (phat - p0) / math.sqrt(p0 * (1 - p0) / n)

p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"標本比率 phat = {phat:.3f}")
print(f"検定統計量 z = {z:.3f}")
print(f"p値 = {p_value:.4f}")

if p_value < alpha:
    print("⇒ 帰無仮説を棄却（有意差あり）")
else:
    print("⇒ 帰無仮説を棄却できない（有意差なし）")
```

```python
import math

mu_0 = 100       # 帰無仮説での母平均
sigma = 3.6     # 母標準偏差 (既知)
n = 12          # 標本サイズ
x_bar = 102.3     # 標本平均
alpha = 0.05     # 有意水準

z_stat = (x_bar - mu_0) / (sigma / math.sqrt(n))

p_value = 2 * (1 - 0.5 * (1 + math.erf(abs(z_stat) / math.sqrt(2))))


# 検定結果
if p_value < alpha:
    print(f"有意水準 {alpha} で帰無仮説を棄却する。")
else:
    print(f"有意水準 {alpha} で帰無仮説を棄却しない。")
```
