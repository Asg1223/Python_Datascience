平均 2.4, 標準偏差 1.1, 標本数 300, 信頼度 95%の区間推定をせよ(正規分布)

```python
import scipy.stats as st
xbar, s, n, alpha = 2.4, 1.1, 300, 0.05 #平均 2.4, 標準偏差 1.1, 標本数 300, 信頼度 95%（1-0.95=0.05） 問題によって適宜変更
z = st.norm.ppf(1 - alpha/2)
ci = (xbar - z * s / (n**0.5), xbar + z * s / (n**0.5))
print(ci)
```

$S^2$ = 0.993, n = 10, 信頼度 95%

```python
import scipy.stats as st
s2, n, alpha = 0.993, 10, 0.05
lower = (n-1)*s2 / st.chi2.ppf(1 - alpha/2, df=n-1)
upper = (n-1)*s2 / st.chi2.ppf(alpha/2, df=n-1)
print(lower, upper)
```

10 人の身長データ：平均 171.98, 分散 77.40「平均が 170 より高い」といえるか (α = 0.05)。

```python
import numpy as np
import scipy.stats as st

# データ（10人の身長）
x = np.array([174.3, 178.0, 176.3, 176.3, 153.8, 178.4, 183.9, 167.5, 164.3, 167.0]) # データは適宜変更

# 検定条件
mu0 = 170       # 帰無仮説 H0: μ = 170
alpha = 0.05    # 有意水準
alternative = "greater"  # 片側検定（平均が170より大きいか）

# t検定（母平均の1標本検定）
t_stat, p_two_sided = st.ttest_1samp(x, popmean=mu0)
n = len(x)
df = n - 1

# 片側検定なので、p値を半分に
p_one_sided = p_two_sided / 2 if t_stat > 0 else 1 - p_two_sided / 2

# 臨界値（右側検定）
t_critical = st.t.ppf(1 - alpha, df)

# 棄却判定
reject = (t_stat > t_critical)

# 結果出力
print(f"t統計量 = {t_stat:.3f}")
print(f"自由度 = {df}")
print(f"p値(片側) = {p_one_sided:.4f}")
print(f"臨界値 t(0.05, {df}) = {t_critical:.3f}")
print("結論: 帰無仮説を棄却する" if reject else "結論: 帰無仮説を棄却しない")
```
